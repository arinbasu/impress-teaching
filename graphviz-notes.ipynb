```python
from graphviz import Digraph
```

```python
dot = Digraph(comment = "a graph", format = 'png')
dot
```

```python
dot.node('A', "this is A")
dot.node('B')
dot.node('L')
dot.edges(['AB', 'AL', 'BL'])
print(dot.source)
dot.render('output/test.gv2')
```
